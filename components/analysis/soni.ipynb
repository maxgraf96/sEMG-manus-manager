```python
import numpy as np

from analysis import load_all_files
```

```python
all_samples, gesture_types, per_recording  = load_all_files(dir="../../user_data")

```

```python
from constants import FEATURE_VECTOR_DIM
import IPython
import librosa

# Take first sample
sample = per_recording[10]
gesture_type = sample[0]
data = sample[1]
print("Sample data shape: ", data.shape)



target_sr = 16000

result = []
# For each channel, sonify the data
for i in range(8):
    channel_data = data[:, i]
    # Resample
    channel_data = librosa.resample(channel_data, orig_sr=200, target_sr=target_sr, res_type="kaiser_best")
    # print("Resampled data shape: ", channel_data.shape)
    channel_data = librosa.effects.pitch_shift(channel_data, sr=target_sr, n_steps=40) # shifted by 4 half steps
    result.append(channel_data)
    
# Average the channels
result = np.mean(result, axis=0)
# Normalise
result = result / np.max(np.abs(result))

# IPython.display.Audio(channel_data, rate=target_sr)
IPython.display.Audio(result, rate=target_sr)
```
